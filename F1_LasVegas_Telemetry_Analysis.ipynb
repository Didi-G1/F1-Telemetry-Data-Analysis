```python
!pip install fastf1
import fastf1  # This brings the official F1 data library into your project
import matplotlib.pyplot as plt  # This library lets us build the line graph

# 1. Load the 2025 Silverstone Qualifying session
session = fastf1.get_session(2025, "Silverstone", "Q")
session.load()  # This downloads the lap times and car speeds automatically

# 2. Extract the absolute fastest single lap for Hamilton and Norris
ham_lap = session.laps.pick_driver("HAM").pick_fastest()
nor_lap = session.laps.pick_driver("NOR").pick_fastest()

# 3. Pull the live car tracking/speed data for those exact laps
ham_telemetry = ham_lap.get_car_data().add_distance()
nor_telemetry = nor_lap.get_car_data().add_distance()

# 4. Create a clean line graph comparing their speed profiles
plt.plot(ham_telemetry["Distance"], ham_telemetry["Speed"], label="Hamilton (Mercedes)", color="cyan")
plt.plot(nor_telemetry["Distance"], nor_telemetry["Speed"], label="Norris (McLaren)", color="orange")

# 5. Label your graph so it looks professional for recruiters
plt.xlabel("Distance around the track (meters)")
plt.ylabel("Car Speed (km/h)")
plt.title("Silverstone 2025: Hamilton vs. Norris Qualifying Speed Profile")
plt.legend()  # Displays the color boxes identifying the drivers
plt.show()  # Renders the finalized graph right on your screen
```

```python
import fastf1  # This brings the official F1 data library into your project
import matplotlib.pyplot as plt  # This library lets us build the line graph

# 1. Load the 2024 Las Vegas Qualifying session
session = fastf1.get_session(2024, "Las Vegas", "Q")
session.load()  # This downloads the lap times and car speeds automatically

# 2. Extract the absolute fastest single lap for Russell & Verstappen
rus_lap = session.laps.pick_driver("RUS").pick_fastest()
ver_lap = session.laps.pick_driver("VER").pick_fastest()

# 3. Pull the live car tracking/speed data for those exact laps
rus_telemetry = rus_lap.get_car_data().add_distance()
ver_telemetry = ver_lap.get_car_data().add_distance()

# 4. Create a clean line graph comparing their speed profiles
plt.plot(rus_telemetry["Distance"], rus_telemetry["Speed"], label="Russell (Mercedes)", color="cyan")
plt.plot(ver_telemetry["Distance"], ver_telemetry["Speed"], label="Verstappen (Red Bull)", color="navy", linestyle="--")

# 5. Label your graph so it looks professional for recruiters
plt.xlabel("Distance around the track (meters)")
plt.ylabel("Car Speed (km/h)")
plt.title("Las Vegas 2024: Russell vs. Verstappen Qualifying Speed Profile")
plt.legend()  # Displays the color boxes identifying the drivers
plt.show()  # Renders the finalized graph right on your screen
```

### Telementary Insights: Las Vegas 2024 (RUS vs. VER)
This project analyzes the qualifying telementary from the 2024 Las Vegas Grand Prix, focusing on the battle for the pole position between George Russell in Mercedes and Max Verstappen in Red Bull.

**Key Findings**
* **Straight-Line Speed:**
    * Both cars reach nearly identical maximum speeds in the longest starights (around 1400m and 5000m), topping out just under 350km/h.
    * Around the 2500m to 3000m mark, the cyan lne (Russell) sits slightly above the dark blue line, showing that the Merceds achived better acceleration out of the oreceding slow corners.
* **Braking and Corner Entry:**
    * Into the heavy braking zones at 300m and 5100m, the dark blue line (Verstappen) drops later than the cyan line. This indicates Verstappen brave later braking points, carrying his top speed a fraction longer.
    * At the deepest point of the heavy braking zones (300,1500m, and 5100m), Russell's minimum speeds drop lower than Verstappen's, showing he sacrified absolute mid-corner speed to focus on a squared-off exit line.
* **Corner Exit Acceleration:**
    * Observe the slopes coming out of the corners (e.g., from 500m to 1000m). The lines are nearly identical, showing both cars had highly optimized traction and engine mpping out of low-speed turns.
